 Databases (SQL or another database-querying language)

 Internet resources that provide data for download

## **Data wrangling**
Data wrangling is the process of preparing the data and getting it into a format that can be used for analysis. The data is often dirty, it requires cleaning before it can be used.

 Human errors: Data is recorded (or even collected) incorrectly

 Computer error: Perhaps we weren't recording entries for a while

 Unexpected values: a question mark for a missing value in a numeric column

 Incomplete information: Think of a survey with optional questions

 Resolution: data collected per second, while we need hourly data

 Format of the data: Data may be recorded in a format that isn't conducive to analysis, which will require us to reshape it

## **Exploratory data analysis (EDA)**
During EDA, we use visualizations and summary statistics to get a better understanding of the data

 Data needs to be prepped before EDA

 Visualizations that are created during EDA may indicate the
need for additional data cleaning.

 Data wrangling uses summary statistics to look for potential data issues, while EDA uses them to understand the data.
