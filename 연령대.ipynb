```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\LG\Desktop\df2.csv")
```

```python
df
```

```python
# 행 인덱스 변경
new_index = ['중구', '동구', '서구', '남구', '북구', '수성구', '달서구', '달성군']
df = df.set_index(pd.Index(new_index))
```

```python
df
```

```python
# 돌리기
df.T
```

```python
# 총인구수 지우기
df =df.T[:101]
```

```python
df
```

```python
# 행 인덱스를 나이로 설정
df.index.name = '나이'
df.reset_index(inplace=True)
```

```python
df
```

```python
# 연령대를 나타내는 라벨
age_labels = ['0-9세', '10-19세', '20-29세', '30-39세', '40-49세', '50-59세', '60-69세', '70-79세', '80-89세', '90-100세']

# 나이에 따른 연령대 라벨 할당
df['연령대'] = pd.cut(df['나이'].astype(int), bins=range(0, 101, 10), labels=age_labels, right=False)

# 연령대 및 지역으로 그룹화하여 각 그룹의 총 인구수 합 계산
result = df.groupby(['연령대']).sum()
```

```python
result = result.drop(columns='나이')
```
